```python
import load_data
import seaborn
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = load_data.get()
```

```python
counts = Counter(df[df["properties_build_year"] != -1]["properties_build_year"])
x = range(min(counts), max(counts))
y = [counts[i] if i in counts else 0 for i in x]
```

```python
fig, ax = plt.subplots()
ax.bar(x, y)

ax.tick_params(axis="y", length=0, pad=10)
ax.tick_params(axis="x", pad=10)
ax.grid(axis="y", linestyle="--")
ax.spines["left"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

ax.annotate('The great artistic\n heritage surge of 1994', (1991, 6.85),
            xytext=(0.7, 0.75), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            fontsize=15, backgroundcolor="white",
            horizontalalignment='right', verticalalignment='top')

plt.savefig('../year-distributions.pdf')
```
